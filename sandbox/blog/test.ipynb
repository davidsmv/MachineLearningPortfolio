```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Data
# -----------------------------
df = pd.DataFrame({
    "Acquisition Method": [
        "Purchase",
        "Adoption",
        "Found on the street",
        "Gift",
        "Other"
    ],
    "Percentage": [15, 29, 20, 33, 3]
})

# -----------------------------
# Sort values
# -----------------------------
df = df.sort_values(by="Percentage", ascending=True)

# -----------------------------
# Create figure
# -----------------------------
fig, ax = plt.subplots(figsize=(9, 5))

# Nice color palette
# colors = plt.cm.Set2.colors[:len(df)]

# Horizontal bars
bars = ax.barh(
    df["Acquisition Method"],
    df["Percentage"],
    # color=colors,
    edgecolor="none"
)

# -----------------------------
# Percentage labels
# -----------------------------
for bar in bars:
    width = bar.get_width()

    ax.text(
        width + 0.5,
        bar.get_y() + bar.get_height()/2,
        f"{width:.0f}%",
        va="center",
        ha="left",
        fontsize=11
    )

# -----------------------------
# Formatting
# -----------------------------
ax.set_title(
    "How Pets Were Acquired in Bogotá (2018)",
    fontsize=15,
    # weight="bold",
    pad=15
)

ax.set_xlabel("Percentage (%)", fontsize=11)
ax.set_ylabel("")

ax.set_xlim(0, 40)

# Grid
# ax.grid(axis="x", linestyle="--", alpha=0.35)

# Remove unnecessary borders
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
# ax.spines["left"].set_visible(False)

# -----------------------------
# Footnote
# -----------------------------
footnote = (
    "Figure elaborated by David Martínez.\n"
    "Note. Data from Instituto Distrital de Protección y Bienestar Animal. (2018). "
    "Caracterización de animales de compañía en Bogotá D.C. \n"
    "https://www.animalesbog.gov.co/documentos/3_CARACTERIZACI%C3%93N_COMPA%C3%91%C3%8DA_2018.pdf"
)

fig.text(
    0.01,
    -0.03,
    footnote,
    ha="left",
    fontsize=8,
    color="dimgray"
)

# Leave room for the footnote
plt.tight_layout(rect=[0, 0.06, 1, 1])

# -----------------------------
# Save figure
# -----------------------------
plt.savefig(
    "pet_acquisition_barh.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Data
# -----------------------------
df = pd.DataFrame({
    "Year": ["April 2025", "April 2026"],
    "Adoptions": [480, 130]
})

before = df.loc[0, "Adoptions"]
after = df.loc[1, "Adoptions"]

drop = (before - after) / before * 100

# -----------------------------
# Figure
# -----------------------------
fig, ax = plt.subplots(figsize=(8, 5))

# Line connecting both years
ax.plot(
    [0, 1],
    [before, after],
    color="lightgray",
    linewidth=3,
    zorder=1
)

# Points
ax.scatter(
    0,
    before,
    s=220,
    color="gray",
    zorder=3
)

ax.scatter(
    1,
    after,
    s=220,
    color="#D55E00",
    zorder=3
)

# Value labels
ax.text(
    0,
    before + 25,
    f"{before}",
    ha="center",
    fontsize=13,
    fontweight="bold"
)

ax.text(
    1,
    after + 25,
    f"{after}",
    ha="center",
    fontsize=13,
    fontweight="bold",
    # color="#D55E00"
)

# Annotation
mid_y = (before + after) / 2

ax.text(
    0.68,              # Slightly right of center
    mid_y + 10,         # Slightly above the middle
    f"↓ {drop:.1f}%\ndecrease",
    ha="left",
    va="center",
    fontsize=13,
    fontweight="bold",
    color="#D55E00",
    bbox=dict(
        facecolor="white",
        edgecolor="none",
        pad=2
    )
)
# X-axis labels
ax.set_xticks([0, 1])
ax.set_xticklabels(["April\n2025", "April\n2026"], fontsize=11)

# -----------------------------
# Style
# -----------------------------
ax.set_title(
    "Pet adoptions in Bogotá fell by nearly 73% in one year",
    fontsize=16,
    # fontweight="bold",
    loc="left",
    pad=20
)

# Remove unnecessary clutter
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

ax.grid(False)
ax.set_yticks([])
ax.tick_params(axis='x', length=0)

# Add some margins
ax.set_xlim(-0.25, 1.25)
ax.set_ylim(0, 560)

footnote = (
    "Figure elaborated by David Martínez.\n"
    "Note. Data from Canal Capital. (2026, April 27). "
    "Bogotá registra caída en adopción de perros y gatos en 2026 | Comunidades [Video]. YouTube. \n"
    "https://www.youtube.com/watch?v=7zXk0KQwGcA"
)

# Footnote
fig.text(
    0.01,
    -0.04,
    footnote,
    fontsize=8,
    color="gray"
)

plt.tight_layout(rect=[0, 0.05, 1, 1])

# Save
plt.savefig(
    "pet_adoptions_slope_chart.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
