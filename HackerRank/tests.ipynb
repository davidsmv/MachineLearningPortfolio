```python
# Project: https://www.hackerrank.com/challenges/predicting-office-space-price/problem?isFullScreen=true
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate sample non-linear data
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
print(X)
y = np.array([1, 4, 9, 16, 25, 36])  # Quadratic pattern (y = x^2)

# ---- CASE 1: Linear Regression without PolynomialFeatures (Wrong Approach) ----
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

# ---- CASE 2: Linear Regression with PolynomialFeatures (Correct Approach) ----
poly = PolynomialFeatures(degree=2)  # Transform features to include X^2
X_poly = poly.fit_transform(X)
print(X_poly)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)

# ---- Plot Results ----
plt.scatter(X, y, color="blue", label="Actual Data")  # Scatter plot of real data
plt.plot(X, y_pred_linear, color="red", linestyle="dashed", label="Linear Regression (Wrong)")
plt.plot(X, y_pred_poly, color="green", label="Polynomial Regression (Correct)")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear vs Polynomial Regression")
plt.show()
```
